```python
%%manim -qh -v WARNING afun
import numpy as np
class afun(Scene):
    def construct(self):
        code_ = """
v = np.array([1,2,3,4])
X = np.reshape(v, (2,2), order = 'F')
A = np.array([
    [2,3],
    [3,5]
])
B = np.array([
    [5,3],
    [7,2]
])
    
result = A @ X + X @ B
    
result = result.flatten('F')
"""
        code = Code(code = code_,language="python", style=Code.styles_list[15]).scale(0.7)
        self.play(Write(code))
        self.wait(1)
        self.play(ApplyMethod(code.to_edge, LEFT))
        v_ = np.array([[1,2,3,4]]).T
        v = Matrix(v_).scale(0.7).move_to(RIGHT*3.5)
        #

        vnxn_ = np.reshape(v_,(2,2), order = 'F')
        vnxn = Matrix(vnxn_).scale(0.7).move_to(v)
        sylvester = VGroup(
            Matrix([[2,3],[3,5]]),
            Matrix(vnxn_),
            MathTex("+"),
            Matrix(vnxn_),
            Matrix([[5,3],[7,2]]),
        ).scale(0.5).arrange(RIGHT).move_to(vnxn)
        result_ = np.array([
            [34, 27],
            [51, 43]
        ])
        resultFlat = np.array([result_.flatten(order = 'F').T])
        result = VGroup(
            Matrix(result_).move_to(vnxn),
            Matrix(resultFlat).move_to(vnxn)
        ).scale(0.7)
        
        self.play(Indicate(code[2][0]))
        self.play(Write(v))
        self.wait
        self.play(Indicate(code[2][1]))
        self.play(ReplacementTransform(v, vnxn))
        self.wait()
        self.play(Indicate(code[2][11]))
        self.play(ReplacementTransform(vnxn, sylvester))
        self.wait()
        self.play(ReplacementTransform(sylvester, result[0]))
        self.wait()
        self.play(Indicate(code[2][13]))
        self.play(ReplacementTransform(result[0], result[1]))
        self.wait(2)
        code_final_ = """
def afun(v):
    #Primero se hace un "reshape" 
    #del vector "v" de dimensi√≥n n^2, 
    #a una matriz X de dimensi√≥n nxn
    X = np.reshape(v, (n,n), order = 'F')
    #Ecuaci√≥n de Sylvester
    out = A @ X + X @ B
    # Se aplica el operador "vec" a la matriz "out" 
    # y se retorna
    return out.flatten('F')
"""
        code_final = Code(code = code_final_,language="python", style=Code.styles_list[15]).scale(0.7).to_edge(LEFT)
        self.play(ReplacementTransform(code, code_final),Unwrite(result[1]))
        self.play(ApplyMethod(code_final.move_to, ORIGIN))

        self.wait(2)
        

    
```
